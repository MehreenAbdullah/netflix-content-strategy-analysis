### Release Trend and Peak Production Year Analysis
<a id=section408></a>

```python
netflix_copy.head(2)
```

```python
# no. of movies released by every year

netflix_copy.groupby(['release_year'])['release_year'].count().sort_values(ascending = False)
netflix_copy.release_year.value_counts().to_frame('Value_count')
```

```python
top_years = netflix_copy.groupby('release_year').size().sort_values(ascending=False).head(10)
top_years = top_years.sort_index()
plt.figure(figsize=(10,6))
bars = plt.bar(top_years.index.astype(str), top_years.values)
plt.title("Top 10 Release Years by Number of Titles", fontsize=14)
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, 
             int(height), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

**Observation:
The highest number of titles were released in 2018, followed closely by 2017 and 2016. 
This indicates that the late 2010s marked a peak production period for content creation. 
The upward trend from 2015 to 2018 suggests accelerated growth in the streaming industry 
before stabilizing slightly in subsequent years.

### Year with Maximum Content Added to Netflix

```python
netflix_copy.dtypes
```

```python
netflix_copy["date_added"] = pd.to_datetime(netflix_copy["date_added"], errors="coerce")
```

```python
year_added_counts = (
    netflix_copy["date_added"]
    .dt.year
    .value_counts()
    .sort_index())
year_added_counts = year_added_counts.loc[2015:]

plt.figure(figsize=(9,5))

bars = plt.bar(
    year_added_counts.index.astype(str),
    year_added_counts.values)

plt.title("Year-wise Content Added to Netflix (2015–2021)", fontsize=13, weight='bold')
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

**Observation:
Content additions increased sharply between 2016 and 2019, indicating an aggressive expansion phase. The peak occurred in 2019, after which content additions slightly declined in 2020 and 2021, suggesting a stabilization period following rapid growth.

### Geographic Concentration of Content Production

```python
country_share = (
    netflix_copy[netflix_copy["country"] != "missing"]["country"]
    .value_counts(normalize=True)
    .head(5) * 100)
country_share.round(2)
```

Observation:
Content production is highly concentrated in the United States, contributing over one-third of total titles. India is the second-largest contributor, while the remaining countries contribute relatively small shares. This indicates a geographically concentrated content sourcing model.

### Content Rating Distribution Across Movies and TV Shows

```python
valid_ratings = [
    "G", "PG", "PG-13", "R",
    "TV-G", "TV-PG", "TV-14", "TV-MA",
    "TV-Y", "TV-Y7", "TV-Y7-FV", "NC-17"]

rating_mix = (
    netflix_copy[netflix_copy["rating"].isin(valid_ratings)]
    .groupby(["type", "rating"])
    .size()
    .unstack(fill_value=0))
```

```python
plt.figure(figsize=(9,4))

sns.heatmap(
    rating_mix,
    annot=True,
    fmt='g',
    cmap="Blues",
    linewidths=0.5)

plt.title("Audience Rating Distribution by Content Type", weight="bold")
plt.xlabel("Rating Category")
plt.ylabel("Content Type")

plt.tight_layout()
plt.show()
```

**Observation:
The rating distribution shows a strong concentration of content in mature categories, particularly TV-MA and TV-14, for both Movies and TV Shows. This indicates a strategic focus on adult and young-adult audiences. Family-oriented categories such as G, PG, and TV-Y represent a relatively smaller portion of the catalog, suggesting that Netflix’s content mix is primarily geared toward mature viewership segments rather than children-focused programming.

## Conclusion

Netflix experienced rapid expansion between 2016 and 2019, with content additions peaking in 2019 before moderating in subsequent years. The catalog is predominantly movie-driven and geographically concentrated, led by the United States and India.

The rating profile shows a clear focus on mature audiences. A significant share of titles fall under TV-MA (18+) and TV-14 (ages 14+), indicating that Netflix primarily targets adult and young-adult segments rather than children-focused programming.
